```python
#how many members in each team

team_splt=emplys['Team'].value_counts()

#persentage splitting with respect to total employees
t=emplys.Team
prsntg_tm=t.value_counts(normalize=True).mul(100).round(1).astype(str)+'%'

#shows in a dataframe
out=pd.DataFrame({"TEAM":team_splt.index,"COUNT":team_splt.values,"PERCENTAGE":prsntg_tm.values})
print(out)

```

### Visualize the report

```python
out.plot(x="TEAM",y="COUNT",kind="bar",color="m")

plt.title("Team V/S Employees in each team")
plt.xlabel("TEAM")
plt.ylabel("COUNT")
plt.show()
```

### which age group most employees belongs

```python
plt.hist(emplys["Age"],bins=10,color='m')
plt.xlabel("age group")
plt.ylabel("employees")
plt.title("AGE DISTRIBUTION OF EMPLOYEES")
plt.show()
```

### under which team and position spending in term of salary high

```python
grp=emplys.groupby(["Team","Position"])["Salary"].sum().reset_index()
high=grp[grp["Salary"]==grp["Salary"].max()]
print("the team and position in term of salary are : ",high)
```

```python
#correlation between age and salary

emplys['Salary']=emplys['Salary'].fillna(0)
emplys['Salary']=emplys['Salary'].astype(int)
emplys['Age'].corr(emplys['Salary'])
plt.title("correlation between age and salary")
sns.scatterplot(x="Age",y="Salary",data=emplys,color="m")
```
